```julia
function count_julia_q()
    h = parsehtml(String(HTTP.get("$baseURL/questions/tagged/julia-lang?sort=newest&pagesize=50").body));
    s1=sel".question-summary"
    s2=sel"[rel=\"next\"]"
    count = 0
    while true
        q = matchall(s1, h.root)
        count = count + length(q)
        elem = matchall(s2, h.root)
        if length(elem) > 0
            nxt = getattr(elem[1], "href")
            h = parsehtml(String(HTTP.get("$baseURL$nxt").body))
        else
            break
        end
    end
    println("Total number of Julia Questions: $count")
end

@time count_julia_q()
```

```julia
const queue = Channel(10)
const count2 = Ref{Int}()

put!(queue, "$baseURL/questions/tagged/julia-lang?sort=newest&pagesize=50")

function count_julia_q_async()
    s1=sel".question-summary"
    s2=sel"[rel=\"next\"]"
    @sync while true
        u = take!(queue)
        if u == nothing; break; end
        @async begin 
            h = parsehtml(String(HTTP.get(u).body));
            q = matchall(s1, h.root)
            count2[] = count2[] + length(q)
            elem = matchall(s2, h.root)
            if length(elem) > 0
                nx = getattr(elem[1], "href")
                put!(queue, "$baseURL$nx")
            else
                put!(queue, nothing)
            end
        end
    end

end

@time count_julia_q_async()
```

```julia
function count_q(tag)
           client = HTTP.Client()
           h1 = parsehtml(String(HTTP.get(client, "$baseURL/questions/tagged/$tag?sort=newest&pagesize=50").body));
           s1=sel".question-summary"
           s2=sel"[rel=\"next\"]"
           count = 0
           client = HTTP.Client()
           while true
               q = matchall(s1, h1.root)
               count = count + length(q)
               elem = matchall(s2, h1.root)
               if length(elem) > 0
                   nx = getattr(elem[1], "href")
                   h1 = parsehtml(String(HTTP.get(client, "$baseURL$nx").body))
               else
                   break
               end
           end
           println("Total number of $tag Questions: $count")
end
```

```julia
@time @sync for i in ["julia-lang", "sencha-touch", "jni"]
    @async count_q(i)
end
```

```julia
function count_qs(tag)
           client = HTTP.Client()
           h1 = parsehtml(String(take!(HTTP.get(client, "$baseURL/questions/tagged/$tag?sort=newest&pagesize=50"))));
           s1=sel".question-summary"
           s2=sel"[rel=\"next\"]"
           count = 0
           client = HTTP.Client()
           while true
            sleep(1)
               q = matchall(s1, h1.root)
               count = count + length(q)
               elem = matchall(s2, h1.root)
               if length(elem) > 0
                   nx = getattr(elem[1], "href")
                   h1 = parsehtml(String(take!(HTTP.get(client, "$baseURL$nx"))))
               else
                   break
               end
           end
           println("Total number of $tag Questions: $count")
end
```

```julia
function count_qs_async(tags)
    @sync for i in tags
    @async count_qs(i)
    end
end
```
